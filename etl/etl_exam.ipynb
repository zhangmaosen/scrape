```python
import ollama
from ollama import Client

client = Client(host='100.103.46.96:11434')
client = Client(host='127.0.0.1:11434')
```

```python
system_prompt = """"""
user_prompt = """
请对以下内容进行修改：
删除所有 HTML 标记（如 <div>、<p>、\n 等），将内容转换为纯文本格式；
删除所有涉及未来预测的句子（包括但不限于对未来营收、净利润、市场份额、增速等的预测，例如 2025-2027 年的数据或目标）；
删除所有风险提示相关的句子（包括但不限于竞争加剧、新市场开拓不及预期、原材料价格上涨、食品安全等风险描述）；
保留其余事实性描述内容，确保修改后的文本逻辑连贯，不提及被删除的部分。以下是需要处理的内容：
'{}'
"""
```

```python
from pydantic import BaseModel, Field

class Output(BaseModel):
    '''删除后的内容'''
    after_deleted :str = Field(..., description="删除后的内容")
```

```python
model = 'gemma3:27b'
model = 'gemma3:12b'
file_name = "fin_qa_3.json"
```

```python
from math import e


with open(file_name, 'w+', encoding='utf-8') as out_f:
    out_f.write("[")
    #需要一个 计数器
    for i, report in enumerate(data):
        print(f"processing {i}\n")
        if i < 596:
            continue
        content = report['content']
        #print(user_prompt.format(content))
        response = client.chat(model=model,
                    messages=[
                        {"role": "user", "content": user_prompt.format(content)},
                    ],
                    format=Output.model_json_schema(),
                    options={
                        "temperature": 0.0,
                        "num_ctx": 4096
                    })
        print(response['message']['content'])
        #print(user_prompt.format(content))
        try:
            qa = json.loads(response['message']['content'])
        except Exception as e:
            print("json.loads error")
            qa = {'after_deleted':''}
        qa = qa | report
        #convert qa to josn
        qa = json.dumps(qa, ensure_ascii=False)
        
        if i != 0:
            out_f.write(",\n")
        out_f.write(qa)
        if i > 1000:
            break
        #break
    out_f.write("]")
```
